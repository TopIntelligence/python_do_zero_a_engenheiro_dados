# Funo Lambda

```python
# Funo tradicional
# Operao soma
def soma(x,y):
    print("Soma:",x+y)
    
x=2
y=3
soma(x,y)
```

```python
SomaLambda = lambda x,y: x+y
SomaLambda(x,y)
```

```python
x = lambda a: a*10
print(x(5))
```

```python
# Funes Def com Lambda
def myfunc(n):
    return lambda a: a*n
mydouble = myfunc(2)
```

```python
mydouble(30)
```

# List Comprehensions

```python
curso = ["Aprenda","Funes","Top", "Intelligence", "Academy"]
```

```python
# Criao tradicional
newlist = []

for x in curso:
    if "Ac" in x:
        newlist.append(x)
print(newlist)
```

```python
newlist = [x for x in curso if "Ac" in x]
print(newlist)
```
